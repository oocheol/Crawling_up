```python
def ok_postcode(post):
        try :
                url = 'https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query=%EC%9A%B0%ED%8E%B8%EB%B2%88%ED%98%B8+%EA%B2%80%EC%83%89'

                driver.get(url)
                # driver1.get(url)
                time.sleep(1)

                
                driver.find_element_by_xpath('//*[@id="search_addrkor"]').click()
                time.sleep(1)
                driver.find_element_by_xpath('//*[@id="search_addrkor"]').send_keys(post)
                time.sleep(1)
                driver.find_element_by_xpath('//*[@id="addrkor_form"]/input[2]').click()
                time.sleep(1)

                postcode = driver.find_element_by_xpath('//*[@id="ds_result"]/div/table/tbody/tr/td[2]/strong').text

                df_post = {'우편번호': postcode}
                df = pd.DataFrame(df_post, index=[0])
                df.set_index('우편번호', inplace=True)
        except :
                try:
                        postcode = driver.find_element_by_xpath('//*[@id="loc-main-section-root"]/section/div/div[2]/div[2]/div/div[1]/div[2]/div[1]/span[1]').text[4:]
                        
                        df_post = {'우편번호': postcode}
                        df = pd.DataFrame(df_post, index=[0])
                        df.set_index('우편번호', inplace=True)

                except :
                        # postcode = driver.find_element_by_xpath('//*[@id="loc-main-section-root"]/section/div/div[2]/div[2]/div/div[1]/div[2]/div[1]/span[1]').text[4:]
                        
                        df_post = {'우편번호': post}
                        df = pd.DataFrame(df_post, index=[0])
                        df.set_index('우편번호', inplace=True)
                
        return df
```

```python
def Crawling_Amazon_review(input_num, remain_num=23):
        i = 2
        url = f'https://worker.mturk.com/?sort=updated_desc&page_number={input_num}'

        driver.get(url)
        # driver1.get(url)
        try :
        # 아이디 로그인
                driver.find_element_by_xpath('//*[@id="ap_email"]').send_keys('oocheol@likelion.org')
                driver.find_element_by_xpath('//*[@id="ap_password"]').send_keys('kaka123!@')
                time.sleep(random.randint(3,5))
                driver.find_element_by_xpath('//*[@id="signInSubmit"]').click()

                # driver1.find_element_by_xpath('//*[@id="ap_email"]').send_keys('oocheol@likelion.org')
                # driver1.find_element_by_xpath('//*[@id="ap_password"]').send_keys('kaka123!@')
                # driver1.find_element_by_xpath('//*[@id="signInSubmit"]').click()

                time.sleep(random.randint(3,5))

                url = f'https://worker.mturk.com/?sort=updated_desc&page_number={input_num}'
                driver.get(url)

                # # 리스트 화면으로 이동
                # driver.find_element_by_xpath('/html/body/div[2]/nav/div/div[1]/div/a').click()
                # # driver1.find_element_by_xpath('/html/body/div[2]/nav/div/div[1]/div/a').click()

                time.sleep(random.randint(3,5))
                
        except :
                url = f'https://worker.mturk.com/?sort=updated_desc&page_number={input_num}'
                driver.get(url)
                time.sleep(random.randint(3,5))
        

        while i < remain_num :
                try :
                        driver.find_element_by_xpath(f'//*[@id="MainContent"]/div[4]/div/div/ol/li[{i}]/div[1]/span[6]/span/a').click()
                        time.sleep(random.randint(3,5))
                
                        # 생성 시간
                        created_time = driver.find_element_by_xpath('/html/body/div[2]/div[2]/div/div[2]/div/div[4]/span[3]/span/span')
                        created_time = created_time.get_attribute('title')
                        created_time_name = created_time.replace('/', '.')
                        created_time_name = created_time_name.replace(':', '.')

                        # 타이틀
                        title = driver.find_element_by_xpath('/html/body/div[2]/div[2]/div/div[1]/div/div/div/div[1]')
                        title = title.get_attribute('title')

                        iframe = driver.find_element_by_xpath(f'//*[@id="MainContent"]/div[5]/div/div/iframe')
                        url = iframe.get_attribute('src')
                        driver.get(url)
                        time.sleep(random.randint(3,5))

                        # review_text
                        review_text = driver.find_element_by_xpath('/html/body').text

                        ob = Screenshot_Clipping.Screenshot()
                        time.sleep(random.randint(3,5))
                        ob.full_Screenshot(driver, save_path=r'./Amazon_img/', image_name=f'Amazon_{created_time_name}_{title}.png')
                        
                        

                        Amazon_M_turk = {'created_time': created_time, 'title': title, 'review_text': review_text}
                        Amazon_M_turk_df = pd.DataFrame(Amazon_M_turk, index=[0])
                        Amazon_M_turk_df.set_index('created_time', inplace=True)
                        globals()[f'Amazon_M_turk_df_{i}'] = Amazon_M_turk_df
                        
                        url = f'https://worker.mturk.com/?sort=updated_desc&page_number={input_num}'
                        driver.get(url)
                        time.sleep(random.randint(3,5))
                        i += 1
                        
                except :
                        Amazon_M_turk = {'created_time': '11/2/2021 02:22pm', 'title': 'Classify Receipt'}
                        Amazon_M_turk_df = pd.DataFrame(Amazon_M_turk, index=[0])
                        Amazon_M_turk_df.set_index('created_time', inplace=True)
                        globals()[f'Amazon_M_turk_df_{i}'] = Amazon_M_turk_df
                        
                        url = f'https://worker.mturk.com/?sort=updated_desc&page_number={input_num}'
                        driver.get(url)
                        time.sleep(random.randint(3,5))
                        i += 1
        

        df = pd.concat([globals()[f'Amazon_M_turk_df_{c}'] for c in range(2, remain_num)])

        return df
```
